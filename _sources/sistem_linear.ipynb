# Sistem Linear

Sistem linear merupakan kumpulan dari dua atau lebih persamaan linear

## sistem linear solusi tunggal

dua garis yang berpotongan

```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan pertama: 2x + 3y = 6
x1_1 = np.linspace(-10, 10, 100)
x1_2 = (-8 + 2*x1_1) / 3

# Persamaan kedua: x - y = 2
x2_1 = np.linspace(-10, 10, 100)
x2_2 = -6 + 4*x2_1

# Menggambar grafik
plt.figure(figsize=(3, 3))
plt.plot(x1_1, x1_2, label='2x + 3y = 8')
plt.plot(x2_1, x2_2, label='4x - y = 6')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Sistem Persamaan Linear')
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

## sistem linear tanpa solusi

garis yang tidak berpotongan dan tidak berhimpit

```python
# Persamaan pertama: x1 + 3x2 = 5
x1_1 = np.linspace(-10, 10, 100)
x1_2 = (-8 + x1_1*2) / -3

# Persamaan kedua: x1 + 3x2 = 0
x2_1 = np.linspace(-10, 10, 100)
x2_2 = (-10 + x2_1*4) / -6

# Menggambar grafik
plt.figure(figsize=(3, 3))
plt.plot(x1_1, x1_2, label='2x1 + 3x2 = 8')
plt.plot(x2_1, x2_2, label='4x1 + 6x2 = 10')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Sistem Persamaan Linear')
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

## sistem linear banyak solusi

garis yang berhimpit

```python
# Persamaan pertama: x1 + 3x2 = 5
x1_1 = np.linspace(-10, 10, 100)
x1_2 = (-8 + x1_1*2) / -3

# Persamaan kedua: x1 + 3x2 = 0
x2_1 = np.linspace(-10, 10, 100)
x2_2 = (-16 + x2_1*4) / -6

# Menggambar grafik
plt.figure(figsize=(3, 3))
plt.plot(x1_1, x1_2, label='2x1 + 3x2 = 8')
plt.plot(x2_1, x2_2, label='4x1 + 6x2 = 16')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Sistem Persamaan Linear')
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
