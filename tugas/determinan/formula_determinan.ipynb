## Ekspansi Kofaktor
Ekspansi Kofaktor adalah salah satu metode mencari determinan suatu matriks, di mana dalam metode ini memanfaatkan kofaktor. Dalam metode ini dikenal istilah Minor Elemen yang ditulis $M_{ij}$ dan kofaktor elemen yang ditulis $c_{ij}$.   

Kita akan mencari determinan dari matriks 4x4 dengan formula determinan ekspansi kofaktor

$
\begin{matrix}
    Matrix A = \begin{bmatrix} 
      1&2&3&3 \\
      2&2&2&2 \\
      3&2&1&2 \\
      3&1&1&1 \\
    \end{bmatrix} \\
  \\
\end{matrix} \\  
$  

$
Menggunakan \space Baris \space 2
$  
$
= a_{21}c_{21} + a_{22}c_{22} + a_{23}c_{23} + a_{24}c_{24} \\
$  
  
$c_{21} = (-1)^{2+1}det(M_{21})$  
  
$c_{22} = (-1)^{2+2}det(M_{22})$  
  
$c_{23} = (-1)^{2+3}det(M_{23})$  
  
$c_{24} = (-1)^{2+4}det(M_{24})$

Kita perlu mencari determian matrix-matrix minor dari matriks A sebagai berikut


**Determinan M21**  
Mencari determinan dari $M_{21}$ dengan Ekspansi kofaktor  

$
\begin{matrix} \\
     M_{21} \space = \begin{bmatrix} 
    2&3&3 \\
    2&1&2 \\
    1&1&1 \\
  \end{bmatrix} \\ \\
\end{matrix} \\
$

$
Menggunakan \space Baris \space 2
$  
$
= m_{21}c_{21} + m_{22}c_{22} + m_{23}c_{23}
$  

$
c_{21} = (-1)^{2+1}det(M_{21 \space dari \space matriks M_{21}})
$  
$
\quad \space = (-1)^{2+1}det(
\begin{matrix}
  \begin{bmatrix}
    3&3 \\
    1&1
  \end{bmatrix}) 
\end{matrix} 
$  
$
\quad \space= -1.0
$  


$
c_{22} = (-1)^{2+2}det(M_{22 \space dari \space matriks M_{21}})
$  
$
\quad \space = (-1)^{2+2}det
\begin{matrix}(
\begin{bmatrix}
    2&3 \\
    1&1
\end{bmatrix}) 
\end{matrix} 
$  
$
\quad \space= 1.-1
$  

$
c_{23} = (-1)^{2+3}det(M_{23 \space dari \space matriks M_{21}})
$  
$
\quad \space = (-1)^{2+3}det
\begin{matrix}(
\begin{bmatrix}
    2&3 \\
    1&1
\end{bmatrix})
\end{matrix} 
$  
$
\quad \space= -1.-1
$  

$
m_{21}c_{21} + m_{22}c_{22} + m_{23}c_{23}
$  
$
= 2.(-1).0 + 1.(1).(-1) + 2.(-1).(-1) 
$  

$
det(M_{21}) \space = 0 - 1 + 2 =  1 
$  
Determinan $M_{21}$ adalah **1**

**Kita akan membuktikanya dengan python apakah benar determinan A = 4**

```python
import numpy as np

A = np.array([[1,2,3,3], [2,2,2,2], [3,2,1,2], [3,1,1,1]])
detA = np.linalg.det(A)
print("A: \n", A)
print()
print("Determinan A : ", round(detA))
print("Hasilnya sama dengan perhitungan manual kita menggunakan formula determinan ekspansi kofaktor")
```
