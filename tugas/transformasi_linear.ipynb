# Transformasi Linear

Transformasi linear, adalah fungsi matematis yang memetakan vektor dari satu ruang ke ruang lainnya, mempertahankan struktur aljabar.

Cara menentukan transformasi linear?  
$\qquad$    1. Pusat titik tidak berubah  
$\qquad$    2. Garis lurus ditetapkan sebagai garis lurus  

  
Secara umum pemetaan T adalah transformasi linier jika  
$\qquad 1. T (u + v) = T(u) + T(v)$  
$\qquad 2. T(au) = a T(u)$  


$
T\left[
    \begin{array}{c} 
        x \\ 
        y \\ 
        z 
    \end{array}
\right] = 
\left[
    \begin{array}{c}
        x+y \\
        x-z 
    \end{array}
\right ] 
     for \space all  
\left[
    \begin{array}{c}
        x \\ 
        y \\ 
        z 
    \end{array}
\right ] 
\in\mathbb{R}^3\nonumber
$  

$
    u = (2, 3, 1) \\
    v = (1, 2, 3) \\
$    
  
**1.** $T (u + v) = T(u) + T(v)$  
    
$
    T(u + v) = 
    \begin{bmatrix}
        3 + 5 \\
        3 - 4 \\
    \end{bmatrix}
    =
    \begin{bmatrix}
        8 \\ 
        -1
    \end{bmatrix}
$  

$
    T(u) + T(v) = 
    \begin{bmatrix}
        5 \\
        1
    \end{bmatrix}
    +
    \begin{bmatrix}
        3 \\
        -2
    \end{bmatrix}
    =
    \begin{bmatrix}
        8 \\
        -1
    \end{bmatrix}
$  

$T(u + v) \space sama \space dengan \space T(u) + T(v)$
  
**2**. $T(au) = a T(u)$
  
a = 2
$
    T(au) =
    \begin{bmatrix}
        4 \\
        6 \\ 
        2
    \end{bmatrix}
    =
    \begin{bmatrix}
        4 + 6 \\
        4 - 2
    \end{bmatrix}
    =
    \begin{bmatrix}
        10 \\
        2
    \end{bmatrix}
$  
  
$
    aT(u) = 
    2
    \begin{bmatrix}
        5 \\
        1 \\
    \end{bmatrix}
    =
    \begin{bmatrix}
        10 \\
        2
    \end{bmatrix}
$

$T(au) \space sama \space dengan \space aT(v)$

Jadi T merupakan transformasi linear

## Matriks refleksi

Matriks refleksi adalah jenis khusus dari transformasi linear dalam geometri yang menghasilkan refleksi (pemantulan) terhadap suatu garis, bidang, atau hiperbidang tertentu. Dalam ruang dua dimensi, matriks refleksi sering kali merepresentasikan pemantulan terhadap sumbu x, sumbu y, atau garis dengan kemiringan tertentu.  

untuk mendapatkan refleksi terhadap sumbu kita dapat melakukan  
   
sumbu $x$   
  
  
$
    R_x =
     \begin{bmatrix}
        1 & 0 \\
         0 & -1 
    \end{bmatrix}
$  
  
$
    \begin{bmatrix}
        1 & 0 \\
         0 & -1 
    \end{bmatrix}
    \begin{bmatrix}
        x_1 \\
        x_2
    \end{bmatrix}
$  

  
sumbu $y$ 
  
$
    R_y =
    \begin{bmatrix}
        -1 & 0 \\
         0 & 1 
    \end{bmatrix}
$
  
$
    \begin{bmatrix}
        -1 & 0 \\
         0 & 1 
    \end{bmatrix}
    \begin{bmatrix}
        x_1 \\
        x_2
    \end{bmatrix}
$

contoh:  

tentukan bayang v terhadap sumbu x
$v = (2, 3)$  

$
    \begin{bmatrix}
        1 & 0 \\
         0 & -1 
    \end{bmatrix}
    \begin{bmatrix}
        2 \\
        3
    \end{bmatrix}
    =
    \begin{bmatrix}
        2 \\
        -3
    \end{bmatrix}
$   
  
jika $v = (2, 3)$ maka bayangan terhadap sumbu x adalah $w = (2, -3)$  
  

Matriks refleksi juga dapat merefleksikan terhadap garis dengan kemiringan tertentu, seperti garis $y=x$. Matriks refleksi terhadap garis tersebut adalah:  
  
$
    R_{x=y} = 
    \begin{bmatrix}
         0 & 1 \\
         1 & 0 
    \end{bmatrix}
$  
  
$
    \begin{bmatrix}
         0 & 1 \\
         1 & 0 
    \end{bmatrix}
    \begin{bmatrix}
        x_1 \\
        x_2
    \end{bmatrix}
$
